```python
Final_Mushroom_df=Mushroom_df.drop(columns=['STALK-SHAPE', 'STALK-SURFACE-ABOVE-RING','STALK-SURFACE-BELOW-RING','STALK-COLOR-ABOVE-RING','STALK-COLOR-BELOW-RING','VEIL-TYPE','VEIL-COLOR','RING-NUMBER','RING-TYPE'])
```

```python
Final_Mushroom_df['EDIBLE OR POISONOUS'] = Final_Mushroom_df['EDIBLE OR POISONOUS'].map({'p':'Poisonous','e':'Edible'})
Final_Mushroom_df['CAP-SHAPE']=Final_Mushroom_df['CAP-SHAPE'].map({'b':'Bell','c':'Conical','x':'Convex','f':'Flat','k':'Knobbed','s':'Sunken'})
Final_Mushroom_df['CAP-SURFACE']=Final_Mushroom_df['CAP-SURFACE'].map({'f':'Fibrous','g':'Grooves','y':'Scaly','s':'Smooth'})
Final_Mushroom_df['BRUISES']=Final_Mushroom_df['BRUISES'].map({'t':'Bruises','f':'No'})
Final_Mushroom_df['ODOR']=Final_Mushroom_df['ODOR'].map({'a':'Almond','l':'Anise','c':'Creosote','y':'Fishy','f':'Foul','m':'Musty','n':'None','p':'Pungent','s':'spicy'})
Final_Mushroom_df['GILL-ATTACHMENT'] = Final_Mushroom_df['GILL-ATTACHMENT'].map({'a':'Attached','d':'Descending','f':'Free','n':'Notched'})
Final_Mushroom_df['CAP-COLOR']=Final_Mushroom_df['CAP-COLOR'].map({'n':'Brown','b':'Buff','c':'Cinnamon','g':'Gray','r':'Green','p':'Pink','u':'Purple','e':'Red','w':'White','y':'Yellow'})
Final_Mushroom_df['GILL-ATTACHMENT']=Final_Mushroom_df['GILL-ATTACHMENT'].map({'a':'Attached','d':'Descending','f':'Free','n':'Notched'})
Final_Mushroom_df['GILL-SPACING']=Final_Mushroom_df['GILL-SPACING'].map({'c':'Close','w':'Crowded','d':'Distant'})
Final_Mushroom_df['GILL-SIZE']=Final_Mushroom_df['GILL-SIZE'].map({'b':'Broad','n':'Narrow'})
Final_Mushroom_df['GILL-COLOR']=Final_Mushroom_df['GILL-COLOR'].map({'k':'Black','n':'Brown','b':'Buff','h':'Chocolate','g':'Gray','r':'Green','o':'Orange','p':'Pink','u':'Purple','e':'Red','w':'White','y':'Yellow'})
Final_Mushroom_df['STALK-ROOT']=Final_Mushroom_df['STALK-ROOT'].map({'b':'Bulbous','c':'Club','u':'Cup','e':'Equal','z':'Rhizomorphs','z':'Rooted'})
Final_Mushroom_df['SPORE-PRINT-COLOR']=Final_Mushroom_df['SPORE-PRINT-COLOR'].map({'k':'Black','n':'Brown','b':'Buff','h':'Chocolate','r':'Green','o':'Orange','u':'Purple','w':'White','y':'Yellow'})
Final_Mushroom_df['POPULATION']=Final_Mushroom_df['POPULATION'].map({'a':'Abundant','c':'Clustered','n':'Numerous','s':'Scattered','v':'Several','y':'Solitary'})
Final_Mushroom_df['HABITAT']=Final_Mushroom_df['HABITAT'].map({'g':'Grasses','l':'Leaves','m':'Meadows','p':'Paths','u':'Urban','w':'Waste','d':'Woods'})
```
