```python
Final_Mushroom_df=Mushroom_df.drop(columns=['STALK-SHAPE', 'STALK-SURFACE-ABOVE-RING','STALK-SURFACE-BELOW-RING','STALK-COLOR-ABOVE-RING','STALK-COLOR-BELOW-RING','VEIL-TYPE','VEIL-COLOR','RING-NUMBER','RING-TYPE','BRUISES'])
```

```python
Final_Mushroom_df['EDIBLE(0) OR POISONOUS(1)'] = Final_Mushroom_df['EDIBLE(0) OR POISONOUS(1)'].map({'p':'0','e':'1'})
Final_Mushroom_df['CAP-SHAPE']=Final_Mushroom_df['CAP-SHAPE'].map({'b':'Bell','c':'Conical','x':'Convex','f':'Flat','k':'Knobbed','s':'Sunken'})
Final_Mushroom_df['CAP-SURFACE']=Final_Mushroom_df['CAP-SURFACE'].map({'f':'Fibrous','g':'Grooves','y':'Scaly','s':'Smooth'})
Final_Mushroom_df['ODOR']=Final_Mushroom_df['ODOR'].map({'a':'Almond','l':'Anise','c':'Creosote','y':'Fishy','f':'Foul','m':'Musty','n':'None','p':'Pungent','s':'spicy'})
Final_Mushroom_df['GILL-ATTACHMENT'] = Final_Mushroom_df['GILL-ATTACHMENT'].map({'a':'Attached','d':'Descending','f':'Free','n':'Notched'})
Final_Mushroom_df['CAP-COLOR']=Final_Mushroom_df['CAP-COLOR'].map({'n':'Brown','b':'Buff','c':'Cinnamon','g':'Gray','r':'Green','p':'Pink','u':'Purple','e':'Red','w':'White','y':'Yellow'})
Final_Mushroom_df['GILL-ATTACHMENT']=Final_Mushroom_df['GILL-ATTACHMENT'].map({'a':'Attached','d':'Descending','f':'Free','n':'Notched'})
Final_Mushroom_df['GILL-SPACING']=Final_Mushroom_df['GILL-SPACING'].map({'c':'Close','w':'Crowded','d':'Distant'})
Final_Mushroom_df['GILL-SIZE']=Final_Mushroom_df['GILL-SIZE'].map({'b':'Broad','n':'Narrow'})
Final_Mushroom_df['GILL-COLOR']=Final_Mushroom_df['GILL-COLOR'].map({'k':'Black','n':'Brown','b':'Buff','h':'Chocolate','g':'Gray','r':'Green','o':'Orange','p':'Pink','u':'Purple','e':'Red','w':'White','y':'Yellow'})
Final_Mushroom_df['STALK-ROOT']=Final_Mushroom_df['STALK-ROOT'].map({'b':'Bulbous','c':'Club','u':'Cup','e':'Equal','z':'Rhizomorphs','z':'Rooted'})
Final_Mushroom_df['SPORE-PRINT-COLOR']=Final_Mushroom_df['SPORE-PRINT-COLOR'].map({'k':'Black','n':'Brown','b':'Buff','h':'Chocolate','r':'Green','o':'Orange','u':'Purple','w':'White','y':'Yellow'})
Final_Mushroom_df['POPULATION']=Final_Mushroom_df['POPULATION'].map({'a':'Abundant','c':'Clustered','n':'Numerous','s':'Scattered','v':'Several','y':'Solitary'})
Final_Mushroom_df['HABITAT']=Final_Mushroom_df['HABITAT'].map({'g':'Grasses','l':'Leaves','m':'Meadows','p':'Paths','u':'Urban','w':'Waste','d':'Woods'})
```

```python
Mushrooms_df=Final_Mushroom_df.drop(columns=['GILL-ATTACHMENT'])
```

```python
Mushrooms_Dummies_df = pd.get_dummies(Mushrooms_df, columns=['CAP-SHAPE','CAP-SURFACE','CAP-COLOR','ODOR','GILL-SPACING','GILL-SIZE','GILL-COLOR','STALK-ROOT','SPORE-PRINT-COLOR','POPULATION','HABITAT'])
```

```python
Mushrooms_Dummies_df
```

```python
import pandas as pd
from path import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
X = Mushrooms_Dummies_df.copy()
X = X.drop(['EDIBLE(0) OR POISONOUS(1)'], axis=1)
X.head()
```

```python
y = Mushrooms_Dummies_df['EDIBLE(0) OR POISONOUS(1)'].ravel()
y[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1000,test_size=0.95)
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
rf_model = RandomForestClassifier(n_estimators=300, random_state=100,max_depth=15,max_features=10) 
```

```python
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
predictions = rf_model.predict(X_test_scaled)
```

```python
predictions
```

```python
cm = confusion_matrix(y_test, predictions)
```

```python
cm_df = pd.DataFrame(
    cm, index=["Actual Edible", "Actual Poisonous"], columns=["Predicted Edible", "Predicted Poisonous"])

cm_df
```

```python
acc_score = accuracy_score(y_test, predictions)
```

```python
acc_score
```

```python
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
#Random Forest
print ("Processing Random Forest")
rf = RandomForestClassifier(n_estimators = 1000, random_state = 1)
rf.fit(X_train, y_train)
y_test_pred = rf.predict(X_test)
print("Random Forest Algorithm Accuracy Score : {:.2f}%".format(rf.score(X_test,y_test)*100))
```
